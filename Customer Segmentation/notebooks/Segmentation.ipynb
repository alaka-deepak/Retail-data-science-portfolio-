### RFM Metrics

RFM Attribute

<li> R(Recency): Number of days since last purchase
<li> F(Frequency) : Number of transactions
<li> M(Monetary) : Total amount of transactions(revenue contribution)

## Monetary

```python
data['TotalPrice']=data['UnitPrice']*data['Quantity']
data.head()
```

```python
m_rfm=data.groupby('CustomerID')['TotalPrice'].sum()
m_rfm=m_rfm.reset_index()
m_rfm.columns=['CustomerID','Monetary']
m_rfm.head()
```

```python
m_rfm.head()
```

### Frequency

```python
f_rfm=data.groupby('CustomerID')['InvoiceNo'].count()
```

```python
f_rfm=f_rfm.reset_index()
f_rfm
```

```python
f_rfm.columns=['CustomerID','Frequency']
```

```python
f_rfm
```

```python
rfm=pd.merge(m_rfm,f_rfm,on='CustomerID',how='inner')
rfm.head()
```

### Date

```python
data['InvoiceDate']
```

```python
min_date=min(data['InvoiceDate'])
max_date=max(data['InvoiceDate'])
print(min_date,max_date)
```

```python
data.dtypes
```

```python
data['Diff_days']=max_date-data['InvoiceDate']
```

```python
r_rfm=data.groupby('CustomerID')['Diff_days'].min()
r_rfm=r_rfm.reset_index()
r_rfm.columns=['CustomerID','Recency']
```

```python
r_rfm.head()
```

```python
r_rfm['Diff_days']=r_rfm['Recency'].dt.days
r_rfm.head()
```

```python
rfm
```

```python
rfm.describe().T
rfm.shape
```

```python
rfm_scores=rfm.copy()
```

```python
rfm_scores.head()
```

```python
rfm_scores['R_score']=pd.qcut(rfm_scores['Recency_x'],5,labels=[5,4,3,2,1])
rfm_scores['F_score']=pd.qcut(rfm_scores['Frequency'],5,labels=[1,2,3,4,5])
rfm_scores['M_score']=pd.qcut(rfm_scores['Monetary'],5,labels=[1,2,3,4,5])
```

```python
rfm_scores.head()
```

# Creating combined RFM Label

```python
rfm_scores['RFM_Score']=rfm_scores['R_score'].astype(str)+rfm_scores['F_score'].astype(str)+rfm_scores['M_score'].astype(str)
```

```python
rfm_scores.head()
```

```python
## RFM score sum

rfm_scores['RFM_Score_sum']=rfm_scores['R_score'].astype(int)+rfm_scores['F_score'].astype(int)+rfm_scores['M_score'].astype(int)
```
